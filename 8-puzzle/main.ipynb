```python
import numpy as np
import hashlib
```

```python
atual = [[1, 2, 3], [-1, 4, 5], [6, 7, 8]]
np.array(atual)
```

```python
def possivel_passo(atual):
  achou_x_y = None
  for x in range(3):
      for y in range(3):
          if atual[x][y] == -1:
              achou_x_y = (x, y)
              break
      if achou_x_y is not None:
          break

  dir = (achou_x_y[0], achou_x_y[1] + 1) if achou_x_y[1] < 2 else None
  esq = (achou_x_y[0], achou_x_y[1] - 1) if achou_x_y[1] > 0 else None
  baixo = (achou_x_y[0] + 1, achou_x_y[1]) if achou_x_y[0] < 2 else None
  cima = (achou_x_y[0] - 1, achou_x_y[1]) if achou_x_y[0] > 0 else None

  posibilidades = [valor for valor in [esq, dir, cima, baixo] if valor]
  return posibilidades, achou_x_y

possivel_passo(atual)
```

```python
def hash_func(atual, hashs):
  atual_str = str(atual).encode('utf-8')
  new_hash = hashlib.sha256(atual_str).hexdigest()
  if new_hash not in hashs:
    hashs.add(new_hash)
    return True
  return False

hash_func(atual, set())
```

```python
def mover(atual, de, para):
  novo = [valor[:] for valor in atual]
  [novo[de[0]][de[1]], novo[para[0]][para[1]]] = [novo[para[0]][para[1]], novo[de[0]][de[1]]]
  return novo

possibilidades, vazio = possivel_passo(atual)
mover(atual, possibilidades[0], vazio), atual
```

```python
def verifica_estado(atual):
  return atual == [[1, 2, 3], [4, 5, 6], [7, 8, -1]]
verifica_estado(atual)
```

```python
global_hashs = set()
def recusao(atuais):
  novos = []
  for atual in atuais:
    possibilidades, vazio = possivel_passo(atual)
    for possibilidade in possibilidades:
      novo_movimento = mover(atual, possibilidade, vazio)
      if hash_func(novo_movimento, global_hashs):
        novos.append(novo_movimento)
        if verifica_estado(novo_movimento):
          print("Estado final encontrado!")
          return True
      else:
        print("Estado já visitado:", novo_movimento)
        
  if recusao(novos):
    return True

if hash_func(atual, global_hashs):
  if verifica_estado(atual):
    print("Estado final encontrado!")
  else:
    recusao([atual])
```
